```python
def outliers_iqr(data, feature):
    x = data[feature]
    quartile_1, quartile_3 = x.quantile(0.25), x.quantile(0.75),
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * 1.5)
    upper_bound = quartile_3 + (iqr * 1.5)
    outliers = data[(x<lower_bound) | (x > upper_bound)]
    cleaned = data[(x>lower_bound) & (x < upper_bound)]
    return outliers, cleaned
```

```python
outliers, cleaned = outliers_iqr(sber_data, 'full_sq')
print(f'Ð§Ð¸ÑÐ»Ð¾ Ð²Ñ‹Ð±Ñ€Ð¾ÑÐ¾Ð² Ð¿Ð¾ Ð¼ÐµÑ‚Ð¾Ð´Ñƒ Ð¢ÑŒÑŽÐºÐ¸: {outliers.shape[0]}')
print(f'Ð ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð¸Ñ€ÑƒÑŽÑ‰ÐµÐµ Ñ‡Ð¸ÑÐ»Ð¾ Ð·Ð°Ð¿Ð¸ÑÐµÐ¹: {cleaned.shape[0]}')
```

```python
def outliers_iqr_mod(data, feature, left=1.5, right=1.5):
    """
    Ð”Ð°Ð²Ð°Ð¹Ñ‚Ðµ Ð½ÐµÐ¼Ð½Ð¾Ð³Ð¾ Ð¼Ð¾Ð´Ð¸Ñ„Ð¸Ñ†Ð¸Ñ€ÑƒÐµÐ¼ Ð½Ð°ÑˆÑƒ Ñ„ÑƒÐ½ÐºÑ†Ð¸ÑŽ outliers_iqr(). 
    Ð”Ð¾Ð±Ð°Ð²ÑŒÑ‚Ðµ Ð² Ð½ÐµÐµ Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ñ‹ left Ð¸ right, ÐºÐ¾Ñ‚Ð¾Ñ€Ñ‹Ðµ Ð·Ð°Ð´Ð°ÑŽÑ‚ Ñ‡Ð¸ÑÐ»Ð¾ IQR Ð²Ð»ÐµÐ²Ð¾ Ð¸ Ð²Ð¿Ñ€Ð°Ð²Ð¾ Ð¾Ñ‚ Ð³Ñ€Ð°Ð½Ð¸Ñ† ÑÑ‰Ð¸ÐºÐ° (Ð¿ÑƒÑÑ‚ÑŒ Ð¿Ð¾ ÑƒÐ¼Ð¾Ð»Ñ‡Ð°Ð½Ð¸ÑŽ Ð¾Ð½Ð¸ Ñ€Ð°Ð²Ð½Ñ‹ 1.5).
    Ð¤ÑƒÐ½ÐºÑ†Ð¸Ñ, ÐºÐ°Ðº Ð¸ Ñ€Ð°Ð½ÑŒÑˆÐµ Ð´Ð¾Ð»Ð¶Ð½Ð° Ð²Ð¾Ð·Ð²Ñ€Ð°Ñ‰Ð°Ñ‚ÑŒ Ð¿Ð¾Ñ‚ÐµÐ½Ñ†Ð¸Ð°Ð»ÑŒÐ½Ñ‹Ðµ Ð²Ñ‹Ð±Ñ€Ð¾ÑÑ‹ Ð¸ Ð¾Ñ‡Ð¸Ñ‰ÐµÐ½Ð½Ñ‹Ð¹ DataFrame.
    """
    x = data[feature]
    quartile_1, quartile_3 = x.quantile(0.25), x.quantile(0.75),
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * left)
    upper_bound = quartile_3 + (iqr * right)
    outliers = data[(x<lower_bound) | (x > upper_bound)]
    cleaned = data[(x>lower_bound) & (x < upper_bound)]
    return outliers, cleaned

outliers, cleaned = outliers_iqr_mod(sber_data, 'full_sq', left=1, right=6)
# print(outliers, cleaned)
```

```python
# Ð°Ð»Ð³Ð¾Ñ€Ð¸Ñ‚Ð¼ Ð¼ÐµÑ‚Ð¾Ð´Ð° z-Ð¾Ñ‚ÐºÐ»Ð¾Ð½ÐµÐ½Ð¸Ñ

def outliers_z_score(data, feature, log_scale=False):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - 3 * sigma
    upper_bound = mu + 3 * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return outliers, cleaned
```

```python
def outliers_z_score_mod(data, feature, log_scale=False, left=3, right=3):
    """
    Ð”Ð°Ð²Ð°Ð¹Ñ‚Ðµ Ñ€Ð°ÑÑˆÐ¸Ñ€Ð¸Ð¼ Ð¿Ñ€Ð°Ð²Ð¸Ð»Ð¾ 3ÐµÑ… ÑÐ¸Ð³Ð¼, Ñ‡Ñ‚Ð¾Ð±Ñ‹ Ð¸Ð¼ÐµÑ‚ÑŒ Ð²Ð¾Ð·Ð¼Ð¾Ð¶Ð½Ð¾ÑÑ‚ÑŒ ÑƒÑ‡Ð¸Ñ‚Ñ‹Ð²Ð°Ñ‚ÑŒ Ð¾ÑÐ¾Ð±ÐµÐ½Ð½Ð¾ÑÑ‚Ð¸ Ð´Ð°Ð½Ð½Ñ‹Ñ….
    Ð”Ð¾Ð±Ð°Ð²ÑŒÑ‚Ðµ Ð² Ñ„ÑƒÐ½ÐºÑ†Ð¸ÑŽ outliers_z_score() Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ñ‹ left Ð¸ right, ÐºÐ¾Ñ‚Ð¾Ñ€Ñ‹Ðµ Ð±ÑƒÐ´ÑƒÑ‚ Ð·Ð°Ð´Ð°Ð²Ð°Ñ‚ÑŒ Ñ‡Ð¸ÑÐ»Ð¾ ÑÐ¸Ð³Ð¼ (ÑÑ‚Ð°Ð½Ð´Ð°Ñ€Ñ‚Ð½Ñ‹Ñ… Ð¾Ñ‚ÐºÐ»Ð¾Ð½ÐµÐ½Ð¸Ð¹) 
    Ð²Ð»ÐµÐ²Ð¾ Ð¸ Ð²Ð¿Ñ€Ð°Ð²Ð¾ ÑÐ¾Ð¾Ñ‚Ð²ÐµÑ‚ÑÑ‚Ð²ÐµÐ½Ð½Ð¾, ÐºÐ¾Ñ‚Ð¾Ñ€Ñ‹Ðµ Ð¾Ð¿Ñ€ÐµÐ´ÐµÐ»ÑÑŽÑ‚ Ð³Ñ€Ð°Ð½Ð¸Ñ†Ñ‹ Ð¼ÐµÑ‚Ð¾Ð´Ð° z-Ð¾Ñ‚ÐºÐ»Ð¾Ð½ÐµÐ½Ð¸Ñ. 
    ÐŸÐ¾ ÑƒÐ¼Ð¾Ð»Ñ‡Ð°Ð½Ð¸ÑŽ Ð¾Ð±Ð° Ð¿Ð°Ñ€Ð°Ð¼ÐµÑ‚Ñ€Ñ‹ Ñ€Ð°Ð²Ð½Ñ‹ 3
    """
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return outliers, cleaned
```
