```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
# print(sns.__version__)

import plotly
import plotly.express as px
# print(plotly.__version__)

import plotly.graph_objs as go
from plotly.offline import iplot, init_notebook_mode
```

## Данные о клиентах банка
#### Столбцы таблицы:
* CustomerId — идентификатор клиента
* Surname — фамилия клиента
* CreditScore — кредитный рейтинг клиента (чем он выше, тем больше клиент брал кредитов и возвращал их)
* Geography — страна клиента (банк международный)
* Gender — пол клиента
* Age — возраст клиента
* Tenure — сколько лет клиент пользуется услугами банка
* Balance — баланс на счетах клиента в банке
* NumOfProducts — количество услуг банка, которые приобрёл клиент
* HasCrCard — есть ли у клиента кредитная карта (1 — да, 0 — нет)
* IsActiveMember — есть ли у клиента статус активного клиента банка (1 — да, 0 — нет)
* EstimatedSalary — предполагаемая заработная плата клиента
* Exited — статус лояльности (1 — ушедший клиент, 0 — лояльный клиент)

```python
churn_data = pd.read_csv('data/churn.csv')
churn_data = churn_data.drop('RowNumber', axis = 1)
churn_data
```

# 1. Лояльность клиентов

```python
exited_cl = 100*(churn_data['Exited'].mean())
part_clients = pd.DataFrame({
    'clients': ['Лояльные', 'Ушедшие'],    
    'part_cl': [100-exited_cl, exited_cl]
})

fig = plt.figure(figsize=(6, 6))
axes = fig.add_axes([0, 0, 1, 1])
axes.set_title('Cоотношение ушедших и лояльных клиентов', fontsize=16)

axes.pie(
    part_clients['part_cl'],
    labels=part_clients['clients'],
    autopct='%.1f%%',
    explode = [0.1, 0]
);
```

**1.1 Доля ушедших клиентов составляет порядка 1/5 от всех клиентов банка. Большая часть  - 80 %, являются лояльными клентами.**

# 10. Распеределение по категоряям клиентов банка

```python
def get_credit_score_cat(credit_score):
    if credit_score >= 300 and credit_score < 500:
        return "Very_Poor"
    elif credit_score >= 500 and credit_score < 601:
        return "Poor"
    elif credit_score >= 601 and credit_score < 661:
        return "Fair"
    elif credit_score >= 661 and credit_score < 781:
        return "Good"
    elif credit_score >= 781 and credit_score < 851:
        return "Excellent"
    elif credit_score >= 851:
        return "Top"
    elif credit_score < 300:
        return "Deep"
    
churn_data['CreditScoreCat'] = churn_data['CreditScore'].apply(get_credit_score_cat)
# churn_data['CreditScoreCat']

pivot_churn = churn_data.pivot_table(
    values='Exited',
    columns='Tenure',
    index='CreditScoreCat',
)

heatmap = sns.heatmap(data=pivot_churn, cmap='YlGnBu')
heatmap.set_title('Тепловая карта лояльности', fontsize=20);
heatmap.set_xlabel('Время пользования услугами банка, лет');
heatmap.set_ylabel('Кредитный рейтинг');
```

**10.1 Как видно из карты категории клиентов, чаще всего уходят клиенты с самым низким кредитным рейтингом пользуясь услугами банка меньше года. Также высокая доля ушедших клиентов со "стажем" 10 лет и имеющими самый низкий рейтинг.**

**10.2 Меньше всего уходят клиенты с хорошей кредитной историей, практически не зависимо от времени пользования услугами банка.**

**10.3 Заметно повышение доли ушедших клиентов в обе стороны шкалы кредитного рейтинга от среднего уровня (Good). Вероятно имеются две противоположные причины: для низкого рейтинга - клиентам тяжело выполнять условия банка, а для высокого ретинга - нет необходимости пользовать услугами этого банка.**
